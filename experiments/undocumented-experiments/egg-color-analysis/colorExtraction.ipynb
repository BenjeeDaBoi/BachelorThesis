```python
import cv2
import numpy as np
import plotly as ply
import time
```

```python
image = cv2.imread("green_egg_example.png")
height, width, dimension = image.shape

startTime = time.time()
rgb_no_kmeans = []

for h in range(height):
  for w in range(width):
    rgb = ""
    for d in range(dimension):
      rgb = str(image[h][w][d]) + ("" if (d) == dimension+1 else ",") + rgb
    print(rgb)
      
print("Duration:", time.time() - startTime)
```
